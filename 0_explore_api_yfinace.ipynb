## yfinance is a unofficial APIs followed by decommissioned of the official version of Yahoo Finance in 2017 due to wide-spread misuse of data.
## Documentation https://pypi.org/project/yfinance/

```python
import pandas as pd
import yfinance as yf

%matplotlib inline
```

```python
# Rechieve price data of ASX200 index from api

ticker = ["^AXJO"]
start = "2012-01-01"
end = "2022-10-10"

data = yf.download(ticker, start=start, end=end)
price = data['Close']
price
```

```python
# present rechived data in line chart
price.plot(figsize=(10,5))
```
